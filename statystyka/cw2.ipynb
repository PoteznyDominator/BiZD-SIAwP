```python
import numpy as np
import pandas as pd
from scipy.stats import describe, bernoulli, binom, poisson, norm
import matplotlib.pyplot as plt
```

```python
values = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

# Obliczenia
mean = np.sum(values * probabilities)
variance = np.sum((values - mean)**2 * probabilities)
kurtosis = np.sum(((values - mean)/variance)**4 * probabilities) - 3
skewness = np.sum(((values - mean)/variance)**3 * probabilities)

print(f"Średnia: {mean}")
print(f"Wariancja: {variance}")
print(f"Kurtoza: {kurtosis}")
print(f"Skośność: {skewness}")
```

```python
p_bernoulli = 0.5  
samples_bernoulli = np.random.choice([0, 1], size=100, p=[1-p_bernoulli, p_bernoulli])

n_binomial = 20  
p_binomial = 0.4  
samples_binomial = np.random.binomial(n_binomial, p_binomial, size=100)

lambda_poisson = 3  
samples_poisson = np.random.poisson(lambda_poisson, size=100)

print("Bernoulli:", samples_bernoulli)
print("Dwumianowy:", samples_binomial)
print("Poisson:", samples_poisson)
```

```python
describe(samples_bernoulli)
```

```python
describe(samples_binomial)
```

```python
describe(samples_poisson)
```

```python
x_bernoulli = np.array([0, 1])
pmf_bernoulli = bernoulli.pmf(x_bernoulli, p_bernoulli)

x_binomial = np.arange(0, n_binomial + 1)
pmf_binomial = binom.pmf(x_binomial, n_binomial, p_binomial)

x_poisson = np.arange(0, 10)
pmf_poisson = poisson.pmf(x_poisson, lambda_poisson)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Wykres dla rozkładu Bernoulliego
axes[0].bar(x_bernoulli, pmf_bernoulli, width=0.4)
axes[0].set_title('Rozkład Bernoulliego')

# Wykres dla rozkładu Dwumianowego
axes[1].bar(x_binomial, pmf_binomial, width=0.4)
axes[1].set_title('Rozkład Dwumianowy')

# Wykres dla rozkładu Poissona
axes[2].bar(x_poisson, pmf_poisson,width=0.4)
axes[2].set_title('Rozkład Poissona')

# Dostosowanie wyglądu
plt.tight_layout()
plt.show()
```

```python
print(f'Rozkład prawdopodobieństwa dla rozkładu Dwumianowego (n=20, p=0.4):')
for k, prawdopodobienstwo in zip(x_binomial, pmf_binomial):
    print(f'P(X={k}) = {prawdopodobienstwo}')
```

```python
sum_pmf = np.sum(pmf_binomial)
print(f'\nSuma prawdopodobieństw: {sum_pmf}')

print ("Suma prawdopobienstwa jest równa 1" if np.isclose(sum_pmf, 1.0) else "Suma nie jest równa 1") 
```

```python
data1 = np.random.normal(1, 2 , 1000)
data_std = np.random.standard_normal(1000)
```

```python
plt.hist(data1, bins=30, density=True, alpha=0.7, color='blue', label='N(1, 2)')
plt.hist(data_std, bins=30, density=True, alpha=0.7, color='orange', label='Standardowy normalny')

x = np.linspace(-5, 3, 100)
y = norm.pdf(x, loc=-1, scale=.5)

plt.plot(x, y, 'r-', label='N(-1, 0.5)')
plt.legend()
plt.xlabel('Wartość')
plt.ylabel('Gęstość prawdopodobieństwa')
plt.title('Histogramy i wykresy gęstości dla różnych rozkładów normalnych')
plt.show()
```
